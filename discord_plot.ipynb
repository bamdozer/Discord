```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the normalized data
file_path = '/Users/bam/Documents/normalized_discord_activity_data.xlsx'
data = pd.read_excel(file_path)

# Convert the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Plotting the data for each user
plt.figure(figsize=(14, 8))
for user in data.columns[1:]:
    plt.plot(data['Date'], data[user], label=user)

plt.xlabel('Date')
plt.ylabel('EXP')
plt.title('EXP Progression Over Time for Each User')
plt.legend(title="Users", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()

# Save the plot as an image file
plot_path = '/Users/bam/Documents/exp_progression_over_time.png'
plt.savefig(plot_path)

plt.show()
```

```python
import plotly.express as px
import pandas as pd

df = pd.read_excel('/Users/bam/Documents/normalized_discord_activity_data_with_chart.xlsx')

#  Plotly
df_long = pd.melt(df, id_vars=['Date'], var_name='Name', value_name='EXP')

# (spline interpolation)
fig = px.line(df_long, x='Date', y='EXP', color='Name', title='EXP Progression Over Time', 
              line_shape='spline')  # 'spline' makes the lines smoother

# Customize 
fig.update_layout(
    title='EXP Progression Over Time',
    xaxis_title='Date',
    yaxis_title='EXP',
    hovermode='x unified',  # Shows all values for a particular date when hovering
    template='plotly_dark',  # Darker background
    legend_title_text='Users',
    margin=dict(l=40, r=40, t=40, b=40)
)

# Sort the hover data by highest value
fig.update_traces(hovertemplate='<b>%{fullData.name}</b>: %{y}<extra></extra>')

# Save the interactive plot as an HTML file
html_file_path = '/Users/bam/Documents/wexp_progression_over_time.html'
fig.write_html(html_file_path)
```
